### Read input

```python
def read_input(file_path):
    with open(file_path, "r") as file:
        steps = [(line[0], int(line[1:].strip())) for line in file.readlines()]
        values = [-y if x == "L" else y for (x, y) in steps]
        return values
```

```python
values = read_input("example.txt")
values
```

### Part 1

```python
count_zero = 0
dial_value = 50
for v in values:
    # Wrap around and constrain to [0, 99]
    dial_value = (dial_value + v) % 100

    if dial_value == 0:
        count_zero += 1

print(f"Number of times 0: {count_zero}")
```

### Part 2

```python
count_zero = 0
dial_value = 50
for v in values:
    direction = 1 if v > 0 else -1
    steps = abs(v)

    # Simulate the dial movement
    for _ in range(steps):
        dial_value = (dial_value + direction) % 100
        if dial_value == 0:
            count_zero += 1

print(f"Number of times 0: {count_zero}")
```
