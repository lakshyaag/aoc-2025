### Read input

```python
def read_input(file_path):
    with open(file_path, "r") as file:
        list_ids = [list(map(int, idx.split("-"))) for idx in file.read().split(",")]
        return list_ids
```

```python
values = read_input("input.txt")
values
```

### Part 1

```python
def check_range_part1(start, end):
    invalid_ids = []
    for i in range(start, end + 1):
        # Check from middle of number
        str_i = str(i)
        mid = len(str_i) // 2

        if str_i[:mid] == str_i[mid:]:
            invalid_ids.append(i)

    return invalid_ids
```

```python
invalid_ids_list = [check_range_part1(*idx) for idx in values]
invalid_ids_list = [item for sublist in invalid_ids_list for item in sublist]
print(f"Sum of invalid IDs: {sum(invalid_ids_list)}")
```

### Part 2

```python
def check_range_part2(start, end):
    invalid_ids = []
    for i in range(start, end + 1):
        str_i = str(i)
        n = len(str_i)

        is_invalid = False
        # Try each length (from 1 to half since we need at least 2 repetitions)
        for unit_len in range(1, n // 2 + 1):
            if n % unit_len == 0:
                unit = str_i[:unit_len]
                if unit * (n // unit_len) == str_i:
                    is_invalid = True
                    break

        if is_invalid:
            invalid_ids.append(i)

    return invalid_ids
```

```python
invalid_ids_list = [check_range_part2(*idx) for idx in values]
invalid_ids_list = [item for sublist in invalid_ids_list for item in sublist]
print(f"Sum of invalid IDs: {sum(invalid_ids_list)}")
```
