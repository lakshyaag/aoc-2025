```python
from rich import print
from collections import Counter
```

### Read input

```python
def read_input(file_path):
    with open(file_path, "r") as file:
        lines = [line.rstrip("\n") for line in file.readlines()]
        return lines
```

```python
values = read_input("example.txt")
```

### Part 1

```python
def part1(values):
    beams = {values[0].index("S")}
    splits = 0

    for line in values[1:]:
        if "^" not in line:
            continue

        next_beams = set()
        for pos in beams:
            if line[pos] == "^":
                splits += 1
                next_beams.add(pos - 1)
                next_beams.add(pos + 1)

            elif line[pos] == ".":
                next_beams.add(pos)

        beams = next_beams

    return splits
```

```python
print(f"Part 1: {part1(values)}")
```

### Part 2

```python
def part2(values):
    beams = Counter({values[0].index("S"): 1})

    for line in values[1:]:
        if "^" not in line:
            continue

        next_beams = Counter()
        for pos, count in beams.items():
            if line[pos] == "^":
                next_beams[pos - 1] += count
                next_beams[pos + 1] += count

            elif line[pos] == ".":
                next_beams[pos] += count

        beams = next_beams

    return sum(beams.values())


print(f"Part 2: {part2(values)}")
```
