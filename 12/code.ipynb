### Read input

```python
def read_input(file_path):
    with open(file_path, "r") as file:
        *shapes, regions = file.read().strip().split("\n\n")

        return shapes, regions
```

```python
shapes, regions = read_input("input.txt")
shapes, regions
```

### Part 1

```python
present_sizes = [c.count("#") for c in shapes]
count = 0
for region in regions.split("\n"):
    w, h, *present_idxs = map(int, region.replace("x", " ").replace(":", "").split(" "))

    if w * h >= sum(n * size for n, size in zip(present_idxs, present_sizes)):
        count += 1

print(count)
```
