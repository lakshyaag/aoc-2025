```python
from rich import print
import numpy as np
from itertools import combinations
```

### Read input

```python
def read_input(file_path):
    with open(file_path, "r") as file:
        lines = list(map(eval, file))
        return lines
```

```python
values = read_input("example.txt")
values
```

### Part 1

```python
def part1(n_iters: int):
    pairs = sorted(
        combinations(values, 2),
        key=lambda x: np.linalg.norm(np.array(x[0]) - np.array(x[1])),
    )

    circuits = {frozenset([box]) for box in values}

    for i, (p, q) in enumerate(pairs):
        c1 = next(c for c in circuits if p in c)
        c2 = next(c for c in circuits if q in c)

        circuits.discard(c1)
        circuits.discard(c2)
        circuits.add(c1.union(c2))

        if i + 1 == n_iters:
            sorted_circuits = sorted(map(len, circuits))
            result = np.prod(sorted_circuits[-3:])
            return result
```

```python
part1(1000)
```
