```python
from rich import print
from itertools import combinations, pairwise
```

### Read input

```python
def read_input(file_path):
    with open(file_path, "r") as file:
        lines = list(map(eval, file))
        return lines
```

```python
values = read_input("example.txt")
values
```

### Part 1

```python
def calculate_area(x: tuple[int, int], y: tuple[int, int]) -> int:
    length = abs(x[0] - y[0]) + 1
    breadth = abs(x[1] - y[1]) + 1
    return length * breadth


rect_areas = sorted(calculate_area(i, j) for i, j in combinations(values, 2))

print(f"Part 1: Largest area is {rect_areas[-1]}")
```

### Part 2

```python
green_rects = list(pairwise(values + [values[0]]))

max_area = 0

for (x, y), (a, b) in combinations(values, 2):
    if x > a:
        x, a = a, x
    if y > b:
        y, b = b, y

    size = (a - x + 1) * (b - y + 1)

    if size > max_area:
        for (p, q), (r, s) in green_rects:
            if p > r:
                p, r = r, p
            if q > s:
                q, s = s, q

            # Check if the green rectangle is inside the red rectangle
            if p < a and q < b and r > x and s > y:
                break
        else:
            max_area = size

print(f"Part 2: Largest area is {max_area}")
```
