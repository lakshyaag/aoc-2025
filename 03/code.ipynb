### Read input

```python
def read_input(file_path):
    with open(file_path, "r") as file:
        lines = [line.strip() for line in file.readlines()]
        return lines
```

```python
values = read_input("example.txt")
values
```

### Part 1

```python
from itertools import combinations


def process_line_part1(line: str):
    return max(int(line[i] + line[j]) for i, j in combinations(range(len(line)), 2))


print(f"Part 1: {sum(process_line_part1(line) for line in values)}")
```

### Part 2

```python
def process_line_part2(line: str):
    line = list(map(int, line))
    selected_batteries = line[-12:]
    for i in range(len(line) - 13, -1, -1):
        current_battery = line[i]
        for j in range(len(selected_batteries)):
            if current_battery >= selected_batteries[j]:
                selected_batteries[j], current_battery = (
                    current_battery,
                    selected_batteries[j],
                )
            else:
                break

    return int("".join(map(str, selected_batteries)))
```

```python
print(f"Part 2: {sum(process_line_part2(line) for line in values)}")
```
